# Two Types of Interfaces

>MATLAB style

>Object Oriented style

```python
# MATLAB style
plt.figure(figsize=(7,4))

plt.subplot(2,1,1)
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)
plt.plot(x, np.cos(x))

plt.show()
```

```python
plt.figure(figsize=(7,4))

plt.subplot(2,2,1)
plt.plot(x, np.sin(x))

plt.subplot(2,2,2)
plt.plot(x, np.cos(x))

plt.subplot(2,2,4)
plt.plot(x, np.tan(x))

plt.show()
```

```python
# Object Oriented style

fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
plt.show()
```

```python
# Object Oriented style

fig, ax = plt.subplots(2,2)

ax[0][0].plot(x, np.sin(x))
ax[1][1].plot(x, np.cos(x))
plt.show()
```

```python
# Object Oriented style

fig, ax = plt.subplots(2,2)

ax[0,0].plot(x, np.sin(x))
ax[1,1].plot(x, np.cos(x))
plt.show()
```

### Adjusting the Plot: Line Colors and Styles

```python
plt.figure(figsize=(5,3))

plt.plot(x, np.sin(x-0), color='blue') # specify color by name
plt.plot(x, np.sin(x-1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x-2), color='0.75') # Grayscale between 0 and 1, 0 will be dark grey and 1 will be white
plt.plot(x, np.sin(x-3), color=(1.0, 0.2, 0.3)) # RGB tuple

plt.show()
```

```python
plt.figure(figsize=(5,3))

plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dotted')
plt.plot(x, x+3, linestyle='dashdot')

plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle=':')
plt.plot(x, x+7, linestyle='-.')

plt.show()
```

```python
plt.figure(figsize=(5,3))

plt.plot(x, x+0, '-g')
plt.plot(x, x+1, '--c')
plt.plot(x, x+2, '-.k')
plt.plot(x, x+3, ':r')

plt.show()
```

```python
plt.figure()
plt.plot(x, np.sin(x-0), color='blue') # specify color by name
plt.show()
```

```python
plt.figure()
plt.axis("equal")
plt.plot(x, np.sin(x-0), color='blue') # specify color by name
plt.show()
```

```python
plt.figure()
plt.plot(x, np.sin(x-0), color='blue') # specify color by name

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(x, np.sin(x), color='blue')

plt.title("A Sine Curve")

plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(x, np.sin(x), color='blue')

plt.title("A Sine Curve")

plt.xlabel("x")
plt.ylabel("sin(x)")

plt.show()
```

```python
plt.figure(figsize=(10,4))

plt.plot(x, np.sin(x), '-b', label='sin(x)')
plt.plot(x, np.cos(x), ':k', label='cos(x)')
plt.legend()

plt.axis("equal")
plt.show()
```

```python
# distplot - It shows the distribution of a Univariate set of observations
# Univariate means One Variable

sns.distplot(tips.total_bill)
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.distplot(tips.total_bill, kde=False)
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.distplot(tips.total_bill, kde=False, bins=30)
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.distplot(tips.total_bill, kde=False, bins=100)
plt.show()
```

```python
# jointplot - It allows you to basically match up two distplots for bivariate analysis

sns.jointplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
plt.show()
```

```python
# pairplot - It will plot pairwise relationships across an entire dataframe(for numerical columns)
# and supports a color hue argument(for categorical columns)

sns.pairplot(tips)
plt.show()
```
